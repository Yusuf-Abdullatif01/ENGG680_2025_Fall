## Part 6: Linear LSE
### (35 marks)

### Curve Fitting

curvefitting.png

### Observations: values of a function f(a) in range of - to , contaminated with random noise

### Unknown States: coefficients of polynomial functions for curve fitting

```python
# step 1
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# step 2
# Load data.xlsx file, which contains the observations (with noise), variance of observations and true observations
# hint: use pandas library to parse excel file and use numpy library to convert it to array
# (1 mark)

df_data = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/ENGG680_2025_Fall/Lab_1_Assignment/data.xlsx') #Loading the Data file
arr_data = df_data.to_numpy() #Converting the Dataframe to a NumPy array

#store the noisy observations, variance of observations and true observations into numpy arrays seperately
# (1 mark)

l = arr_data[:,0] #Noisy Observations as the first column
l_var = arr_data[:,1] #Variance of Observation as the second column
l_true = arr_data[:,2] #True Observations as the third column
```

```python
# step 3
# define the number of observation n
# (1 mark)

n = len(l) # l is the noisy Observation array

# define the observation weight matrix P for LSE as identity matrix
# identity weight matrix simply means the observations will be unweighted in the following LSE
# (1 mark)
P = np.eye(n)  #Identity Matrix (n x n)

# define the range of function input(i.e., constants a), from start to end; and the numerical array of a
start = -1* np.pi
end = np.pi
a = np.linspace(start, end, n)
```

```python
# step 4
# use the matplotlib to plot the true and noisy observations together in one single figure
# use scatter for noisy observation plotting
# for the true observation, label it as "f(a)"" and set color as 'k'
# for the noisy observation scatter plot, label it as "f(a) + e", set color as 'tab:blue', set marker as 'o' and set alpha as 0.3
# include proper axis labels, legend and title
# scale the range of x-axis to be -1 to 1
# keep this format for the rest of the plots in this lab assignment
# (2 marks)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(14, 5)) #Plot True and Noisy Observations together

ax.scatter(a, l, label='f(a) + e', color='tab:blue', marker='o', alpha=0.3) #Scatter plot for Noisy Observations, label it as "f(a) + e", set color as 'tab:blue', set marker as 'o' and set alpha as 0.3

ax.plot(a, l_true, label='f(a)', color='k') #Line plot for True Observations, label it as "f(a)"" and set color as 'k'

ax.set_xlabel('a') #X Axis Label
ax.set_ylabel('l') #Y Axis Label
ax.set_title('Observations Fitting') #Title
ax.legend() #Legend

ax.set_xlim(-1, 1) #scale the range of x-axis to be -1 to 1

ax.grid(True, which='both',ls='-',color='0.65')
fig.tight_layout()
plt.show() #Showing the plot
```

```python
# step 5
# define a curve fitting function named fit_polynomial_curve
# the function should implement the linear least square estimation
# the function should include all important intermediate steps in LSE
# the input and output of the funciton are described in the comment below in red.
# please keep code clean and well-commented
# (7 marks)

#Fitting a Polynomial curve to Noisy Observations using the Linear Least Square Estimation
def fit_polynomial_curve(a: np.ndarray, l: np.ndarray, P: np.ndarray, deg: int) -> tuple:
    '''
    Args:
        a (np.ndarray): function input (constants)
        l (np.ndarray): noisy observations
        P (np.ndarray): observation weight matrix
        deg (int): degree of polynomial to fit to data
    Returns:
        x (np.ndarray): estimated polynomial coefficients
        cov_x (np.ndarray): uncertainty in estimated coefficients
        apv (float): a posterior variance factor (goodness of fit)
        v (np.ndarray): observation residuals
    '''

    # Build the design matrix A [1, a_i, a_i^2, ...]
    A = np.vander(a, deg + 1, increasing=True)

    # Weighted least squares estimation
    AtP = A.T @ P
    N = AtP @ A
    x = np.linalg.solve(N, AtP @ l)  # estimated coefficients

    # Residuals (difference between observed and model predictions)
    v = l - A @ x

    # A posterior variance factor (goodness of fit)
    apv = (v.T @ P @ v) / (len(l) - (deg + 1))

    # Covariance matrix of uncertainty in estimated coefficients
    cov_x = apv * np.linalg.inv(N)

    return (x, cov_x, apv, v)
```

```python
# step 6
# call the fit_polynomial_curve function and fetch the results
# use polynomial degrees of: 2, 4, 6
# (1 marks)

degrees = [2, 4, 6] #listing the degrees of the polynomial complexity
results = [fit_polynomial_curve(a, l, P, deg) for deg in degrees] #calling the fit_polynomial_curve function and fetching the results

#results will contain tuples with Estimated Coefficients, Covariance Matrix, A posterior variance factor and Residuals (from step 5)
```

```python
# step 7
# use the estimated coefficients x and constants array a, construct polynomial functions for each degree
# hint: use polyval function from numpy library (check input argument carefully)
# (1 marks)

# Extract coefficients from results for different polynomial degrees
x2, _, _, _ = results[0]  # Degree 2
x4, _, _, _ = results[1]  # Degree 4
x6, _, _, _ = results[2]  # Degree 6

# Construct polynomial functions for each degree (unweighted)
quadratic_uw = np.poly1d(x2[::-1])  # degree 2
quartic_uw = np.poly1d(x4[::-1])    # degree 4
hexic_uw = np.poly1d(x6[::-1])      # degree 6

# Note:
# np.poly1d expects coefficients in order: [a_n, a_(n-1), ..., a_0]
# Our 'x' coefficients are in ascending order (lowest to highest),
# so we reverse them using [::-1].
```
